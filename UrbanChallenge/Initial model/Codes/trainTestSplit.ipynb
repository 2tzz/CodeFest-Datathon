```python
import os

image_folder = './Datathoon'      # your images folder
label_folder = './Datathoon'      # your labels folder (or same as images folder if labels stored there)
exts = ('.jpg', '.png')        # image extensions

# Make sure label folder exists
os.makedirs(label_folder, exist_ok=True)

# List all image files
image_files = [f for f in os.listdir(image_folder) if f.lower().endswith(exts)]

for img_file in image_files:
    label_file = os.path.splitext(img_file)[0] + '.txt'
    label_path = os.path.join(label_folder, label_file)
    # If label file does NOT exist, create an empty one
    if not os.path.exists(label_path):
        open(label_path, 'w').close()  # create empty file
        print(f'Created empty label file: {label_file}')
```

```python
import os, random, shutil

# Paths
all_images = "./Datathoon"   # Folder containing ALL images & txt labels
train_images = "dataset/images/train"
val_images = "dataset/images/val"
train_labels = "dataset/labels/train"
val_labels = "dataset/labels/val"

# Create directories
os.makedirs(train_images, exist_ok=True)
os.makedirs(val_images, exist_ok=True)
os.makedirs(train_labels, exist_ok=True)
os.makedirs(val_labels, exist_ok=True)

# Get image list
images = [f for f in os.listdir(all_images) if f.lower().endswith((".jpg", ".png"))]
random.shuffle(images)

# Split
split_idx = int(0.8 * len(images))
train_files = images[:split_idx]
val_files = images[split_idx:]

def copy_pair(img_file, dst_img_dir, dst_label_dir):
    # Image path
    src_img_path = os.path.join(all_images, img_file)
    dst_img_path = os.path.join(dst_img_dir, img_file)

    # Label path (same name, .txt extension)
    label_file = os.path.splitext(img_file)[0] + ".txt"
    src_label_path = os.path.join(all_images, label_file)
    dst_label_path = os.path.join(dst_label_dir, label_file)

    # Copy files
    shutil.copy(src_img_path, dst_img_path)
    shutil.copy(src_label_path, dst_label_path)

# Copy train set
for f in train_files:
    copy_pair(f, train_images, train_labels)

# Copy val set
for f in val_files:
    copy_pair(f, val_images, val_labels)

print(f"Train: {len(train_files)} images")
print(f"Val: {len(val_files)} images")
```
